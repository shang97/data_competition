python 中的 split 方法只能通过指定的某一个字符分割字符串，因此要实现**多符号同时分割**的话，就要调用更加强大的 re 模块中的 split() 函数  
参考：cnblogs.com/w-j-c/p/10776527.html

```python
import re

# 每个分隔符用 | 进行分隔， . ? / 需要加转义符
train['a'] = train.text.apply(lambda x: re.split('3750|900|648', x))
```

#### 统计每类新闻中出现次数对多的字符
分组函数的使用 https://www.cnblogs.com/lemonbit/p/6810972.html  
https://www.cnblogs.com/Summer-skr--blog/p/11705892.html

```python
groups = train.groupby('label')
```

```python
# 查看所有分组（name和index）
groups.groups
```

```python
from collections import Counter

for group_name, group in groups:   
    # list(group.text) 只有一个 []
    word_str = ' '.join(list(group.text))
    word_dict = Counter(word_str.split(' '))
    # sorted作用于字典上
    word_count = sorted(word_dict.items(), key=lambda x: x[1], reverse=True)
    
    print(group_name, '\t', word_count[:5])
```
