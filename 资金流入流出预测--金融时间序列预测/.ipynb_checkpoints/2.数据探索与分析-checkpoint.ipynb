注，此处所有的金额单位都是分（百分之一元），例如用户某日买入100万，代表其购买了1万

```python
import warnings 
warnings.filterwarnings('ignore')

import pandas as  pd
import numpy as np
import datetime

import seaborn as sns
import matplotlib.pyplot as plt
# 汉字显示：https://www.jianshu.com/p/d28ae68403d3
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
%matplotlib inline

from scipy import stats
```

```python
%%time
# Load the balance data
data_balance = pd.read_csv('data/user_balance_table.csv')

# add tiemstamp to dataset
data_balance['date'] = pd.to_datetime(data_balance['report_date'], format= "%Y%m%d")
data_balance['day'] = data_balance['date'].dt.day
data_balance['month'] = data_balance['date'].dt.month
data_balance['year'] = data_balance['date'].dt.year
data_balance['week'] = data_balance['date'].dt.week
data_balance['weekday'] = data_balance['date'].dt.weekday

# total amount
total_balance = data_balance.groupby(['date'])['total_purchase_amt','total_redeem_amt'].sum()
total_balance.reset_index(inplace=True)

# Add time stamp
total_balance['day'] = total_balance['date'].dt.day
total_balance['month'] = total_balance['date'].dt.month
total_balance['year'] = total_balance['date'].dt.year
total_balance['week'] = total_balance['date'].dt.week
total_balance['weekday'] = total_balance['date'].dt.weekday

# Load user's information
users = pd.read_csv('data/user_profile_table.csv')
```

### 一、分析大小额用户

#### 获得大额用户的集合

前面的分析可以看出，某些用户的交易额很大，对于日交易量（的影响？）很明显，这里统计四月份后依旧活跃的大额用户（购买量和赎回量都要大于 1e6）

```python
temp = data_balance[(data_balance['total_purchase_amt'] >= 1e6) | (data_balance['total_redeem_amt'] >= 1e6)]
big_users_set = set(temp[temp['date'] >= datetime.datetime(2014,4,1)]['user_id'])
```

大额用户占3903个，占总数的13%

```python
len(big_users_set),  max(data_balance['user_id'])
```

#### 标记大额用户

isin函数

```python
data_balance['big_user'] = 0
data_balance.loc[data_balance['user_id'].isin(big_users_set), 'big_user'] = 1
```

#### 分析大额用户与小额用户的日总交易额的区别

groupby函数的灵活使用

```python
total_balance_big_small = data_balance.groupby(['date','big_user'], as_index=False)['total_purchase_amt','total_redeem_amt'].sum()
```

#### 大额用户与小额用户的每日总交易量图

```python
fig = plt.figure(figsize=(20,8))
plt.title("The time series of big and small user of Purchase and Redeem")

# 大额
plt.plot(total_balance_big_small[total_balance_big_small['big_user'] == 1]['date'], 
         total_balance_big_small[total_balance_big_small['big_user'] == 1]['total_purchase_amt'], label='big_purchase')

plt.plot(total_balance_big_small[total_balance_big_small['big_user'] == 1]['date'], 
         total_balance_big_small[total_balance_big_small['big_user'] == 1]['total_redeem_amt'], label='big_redeem')


# 小额
plt.plot(total_balance_big_small[total_balance_big_small['big_user'] == 0]['date'], 
         total_balance_big_small[total_balance_big_small['big_user'] == 0]['total_purchase_amt'], label='small_purchase')

plt.plot(total_balance_big_small[total_balance_big_small['big_user'] == 0]['date'], 
         total_balance_big_small[total_balance_big_small['big_user'] == 0]['total_redeem_amt'], label='small_redeem')

plt.legend(loc='best')
plt.xlabel("Time")
plt.ylabel("Amount")
plt.show()
```

#### 统计大额小额用户购买量占比

```python
big_all_purchase = np.sum(total_balance_big_small[total_balance_big_small['big_user'] == 1]['total_purchase_amt']) 
small_all_purchase = np.sum(total_balance_big_small[total_balance_big_small['big_user'] == 0]['total_purchase_amt'])
big_all_purchase / small_all_purchase
```

#### 统计大额小额用户赎回量占比

```python
big_all_redeem = np.sum(total_balance_big_small[total_balance_big_small['big_user'] == 1]['total_redeem_amt'])
small_all_redeem = np.sum(total_balance_big_small[total_balance_big_small['big_user'] == 0]['total_redeem_amt'])
big_all_redeem / small_all_redeem
```

#### 画出非0交易（direct_purchase、direct_redeem）的分布图

```python
frequency = data_balance[(data_balance['direct_purchase_amt'] != 0) | 
                         (data_balance['total_redeem_amt'] != 0)][['user_id','tBalance']].groupby('user_id', as_index=False).count()
sns.distplot(frequency['tBalance'])
```

#### 标记出频繁交易用户（tBalance大于30）

```python
# 获取频繁交易用户的集合
hot_users_set = set(frequency[frequency['tBalance'] > 30]['user_id'])

data_balance['is_hot_users'] = 0
data_balance.loc[data_balance['user_id'].isin(hot_users_set) , 'is_hot_users'] = 1
```

```python
total_balance_hotNcold = data_balance.groupby(['date','is_hot_users'], as_index=False)['total_purchase_amt','total_redeem_amt'].sum()
```

#### 频繁交易用户与非频繁交易用户的总购买、赎回量的时序图

```python
fig = plt.figure(figsize=(20,8))
plt.title("The time series of big and small user of Purchase and Redeem")

# 频繁交易
plt.plot(total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 1]['date'], 
         total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 1]['total_purchase_amt'], label='hot_purchase')

plt.plot(total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 1]['date'], 
         total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 1]['total_redeem_amt'], label='hot_redeem')

# 非频繁交易
plt.plot(total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 0]['date'], 
         total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 0]['total_purchase_amt'], label='cold_purchase')

plt.plot(total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 0]['date'], 
         total_balance_hotNcold[total_balance_hotNcold['is_hot_users'] == 0]['total_redeem_amt'], label='cold_redeem')

plt.legend(loc='best')
plt.xlabel("Time")
plt.ylabel("Amount")
plt.show()
```

| 和 & 的结果是不一致的，前者是取并集（但不是相加），后者是取交集 ！！！！

```python
temp = data_balance[['year','month','user_id',
                     'total_purchase_amt','total_redeem_amt']].groupby(['year','month','user_id'],  as_index=False).sum()
user_old_set = set()
plt.figure(figsize=(10,30))
for i in range(1, 9):
    # 14 年每个月的新用户 id
    month_user_set = set(temp[(temp['year'] == 2014) & (temp['month'] == i)]['user_id'])
    this_month = data_balance[(data_balance['year'] == 2014) & (data_balance['month'] == i)]
    
    # 标记处每个月的新用户
    this_month['cold'] = 0
    # 集合减法，只取该月新出现的用户 id
    this_month.loc[this_month['user_id'].isin(month_user_set - user_old_set), 'cold'] = 1
    
    # 第一个月都是该月的新用户，所以箱线图只有一个
    plt.subplot(4,2,i)
    plt.title('Month : ' + str(i))
    # 取并集，有一个不是 0 就行
    sns.boxplot(x="cold", y="total_purchase_amt" , data=this_month[(this_month['direct_purchase_amt'] != 0) | 
                                                                   (this_month['total_redeem_amt'] != 0)])
    user_old_set = user_old_set | month_user_set
plt.show()
```
